```python
# Vanilla Fully Connected versus general

plt.style.use('seaborn-deep')
kwargs = dict(color=['hotpink', 'cyan', 'teal'], alpha=0.9, bins=24)#, density=True)

fig, ax = plt.subplots(constrained_layout=False)
ax.grid(alpha=0.4)

fig.tight_layout()
fig.subplots_adjust(left=0.19, bottom=0.09, right=0.98, top=0.92, wspace=0.17, hspace=0.17)

x1=general['dups'].loc[general['dups'] <= 24]
x2=vanillaGeneral['dups'].loc[vanillaGeneral['dups'] <= 24]
x6=squelchingGeneral['dups'].loc[squelchingGeneral['dups'] <= 24]
bins = 24

plt.gca().set(title='Distribution of Replicated Messages', ylabel='Number of Messages', xlabel="Number of Replicas")

plt.rcParams.update({'font.size': 18})
plt.hist([x1, x2, x6], **kwargs, label=['GS 1 topic', 'Vanilla', 'Squelching'])
plt.legend(loc='upper center', bbox_to_anchor=(0.58, 1))
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.xticks(fontsize=18)
plt.yticks(fontsize=16)

fig.set_size_inches(11, 4.3)

# ax.annotate('', xy=(22.7, 6650),  xycoords='data',
#             xytext=(0.95, 0.9), textcoords='axes fraction',
#             arrowprops=dict(facecolor='red', shrink=0.05),
#             horizontalalignment='left', verticalalignment='top',
#             )

fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralAbsolute2.pdf', format='pdf', facecolor='white', edgecolor='none', bbox_inches='tight', dpi=600)
fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralAbsolute2.png', format='png', facecolor='white', edgecolor='none', bbox_inches='tight', dpi=600)

plt.show()
```

```python
#Imports and everything else, better to not modify this part
import numpy as np
import sys
import pandas as pd
#
import matplotlib.pyplot as plt

import matplotlib.ticker as ticker
# from collections import namedtuple

try:
    # for Python2
    from Tkinter import *   ## notice capitalized T in Tkinter 
except ImportError:
    # for Python3
    from tkinter import *   ## notice lowercase 't' in tkinter here
    
    
    pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('/home/flav/gossip/gossip_data/dupsGeneral.csv')#, usecols=[0,2,5])
general = pd.DataFrame(data)

data = pd.read_csv('/home/flav/gossip/gossip_data/dupsvanillaGeneral.csv')#, usecols=[0,2,5])
vanillaGeneral = pd.DataFrame(data)

data = pd.read_csv('/home/flav/gossip/gossip_data/dupsvanillaValidator.csv')#, usecols=[0,2,5])
vanillaValidator = pd.DataFrame(data)

data = pd.read_csv('/home/flav/gossip/gossip_data/dupsValidator.csv')#, usecols=[0,2,5])
validator = pd.DataFrame(data)

data = pd.read_csv('/home/flav/gossip/gossip_data/dupsUNL.csv')#, usecols=[0,2,5])
unl = pd.DataFrame(data)
```

```python

plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
kwargs = dict(alpha=0.5, bins=50, density=True, stacked=True)

x1=general['dups']
x2=vanillaGeneral['dups']
x3=validator['dups']
x4=vanillaValidator['dups']
x5=unl['dups']

# Plot
plt.hist(x1, **kwargs, color='g', label='General')
plt.hist(x2, **kwargs, color='b', label='VanillaGeneral')
plt.hist(x3, **kwargs, color='r', label='Validator')
plt.hist(x4, **kwargs, color='y', label='VanillaValidator')
plt.hist(x5, **kwargs, color='c', label='UNL')
# plt.hist(x4, **kwargs, color='y', label='UNL')
plt.gca().set(title='Distribution of Duplicates', ylabel='Distribution')
# plt.xlim(50,75)
plt.legend();

```

```python
# Vanilla Fully Connected versus general

plt.style.use('seaborn-deep')
kwargs = dict(color=['hotpink', 'cyan'], alpha=0.9, bins=24)#, density=True)

fig, ax = plt.subplots(constrained_layout=False)
ax.grid(alpha=0.4)

fig.tight_layout()
fig.subplots_adjust(left=0.19, bottom=0.09, right=0.98, top=0.92, wspace=0.17, hspace=0.17)

x1=general['dups']
x2=vanillaGeneral['dups']
bins = 24

plt.gca().set(title='Distribution of Duplicates (Absolute)', ylabel='Number of Messages', xlabel="Number of Duplicates")

plt.hist([x1, x2], **kwargs, label=['GossipSub 1 For All Validations', 'Vanilla Fully Connected Mesh'])
plt.legend(loc='upper center')

fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralAbsolute.pdf', format='pdf', facecolor='white', edgecolor='none', bbox_inches='tight')
fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralAbsolute.png', format='png', facecolor='white', edgecolor='none', bbox_inches='tight')

plt.show()
```

```python
# Vanilla Fully Connected versus general

plt.style.use('seaborn-deep')
kwargs = dict(color=['hotpink', 'cyan'], alpha=0.9, bins=24, density=True)

fig, ax = plt.subplots(constrained_layout=False)
ax.grid(alpha=0.4)

fig.tight_layout()
fig.subplots_adjust(left=0.19, bottom=0.09, right=0.98, top=0.92, wspace=0.17, hspace=0.17)

x1=general['dups']
x2=vanillaGeneral['dups']
bins = 24

plt.gca().set(title='Distribution of Duplicates (Density)', ylabel='Density', xlabel="Number of Duplicates")

plt.hist([x1, x2], **kwargs, label=['GossipSub 1 For All Validations', 'Vanilla Fully Connected Mesh'])
plt.legend(loc='upper center')

fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralDensity.pdf', format='pdf', facecolor='white', edgecolor='none', bbox_inches='tight')
fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesGeneneralDensity.png', format='png', facecolor='white', edgecolor='none', bbox_inches='tight')

plt.show()
```

```python
# Vanilla Max Deegree 16 versus validator versus UNL

plt.style.use('seaborn-deep')
kwargs = dict(color=['mediumvioletred', 'skyblue', 'teal'], alpha=0.9, bins=16, histtype='bar')#, density=True)

fig, ax = plt.subplots(constrained_layout=False)
ax.grid(alpha=0.4)

fig.tight_layout()
fig.subplots_adjust(left=0.19, bottom=0.09, right=0.98, top=0.92, wspace=0.17, hspace=0.17)

x1=validator['dups'].loc[validator['dups'] <= 16]
x2=vanillaValidator['dups'].loc[vanillaValidator['dups'] <= 16]
x3=unl['dups'].loc[unl['dups'] <= 16]

plt.gca().set(title='Distribution of Duplicates (Absolute)', ylabel='Number of Messages', xlabel="Number of Duplicates")


plt.hist([x1, x3, x2], **kwargs, label=['GossipSub 1 topic/validator', 'GossipSub 1 topic/UNL', 'Vanilla Mesh Avg Deegree 16'])
plt.legend(loc='upper center')


fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesValidatorAbsolute.pdf', format='pdf', facecolor='white', edgecolor='none', bbox_inches='tight')
fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesValidatorAbsolute.png', format='png', facecolor='white', edgecolor='none', bbox_inches='tight')

plt.show()
```

```python
# Vanilla Max Deegree 16 versus validator versus UNL

plt.style.use('seaborn-deep')
kwargs = dict(color=['mediumvioletred', 'skyblue', 'teal'], alpha=0.9, bins=16, histtype='bar', density=True)

fig, ax = plt.subplots(constrained_layout=False)
ax.grid(alpha=0.4)

fig.tight_layout()
fig.subplots_adjust(left=0.19, bottom=0.09, right=0.98, top=0.92, wspace=0.17, hspace=0.17)

x1=validator['dups'].loc[validator['dups'] <= 16]
x2=vanillaValidator['dups'].loc[vanillaValidator['dups'] <= 16]
x3=unl['dups'].loc[unl['dups'] <= 16]

plt.gca().set(title='Distribution of Duplicates (Density)', ylabel='Density', xlabel="Number of Duplicates")


plt.hist([x1, x3, x2], **kwargs, label=['GossipSub 1 topic/validator', 'GossipSub 1 topic/UNL', 'Vanilla Mesh Avg Deegree 16'])
plt.legend(loc='upper center')


fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesValidatorDensity.pdf', format='pdf', facecolor='white', edgecolor='none', bbox_inches='tight')
fig.savefig('/home/flav/gossip/gossip_data/figures/DuplicatesValidatorDensity.png', format='png', facecolor='white', edgecolor='none', bbox_inches='tight')

plt.show()
```
